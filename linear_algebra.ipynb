```python
import numpy as np
```

```python
b = np.array([1, 2, 3])
M = np.array([[2, -1, 0], [-1, 2, -1], [0, -1, 2]])
x = np.linalg.solve(M, b)
res = b - np.dot(M, x)
```

```python
np.dot(x, b)
```

```python
np.dot(b, res)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vector b
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='r', label='b')

# Plot vector x
ax.quiver(0, 0, 0, x[0], x[1], x[2], color='b', label='x')

# Plot res
# scale res to make it visible
res = res * 1e10
print(res)
ax.quiver(0, 0, 0, res[0], res[1], res[2], color='g', label='res')

ax.set_xlim([0, max(b[0], x[0])])
ax.set_ylim([0, max(b[1], x[1])])
ax.set_zlim([0, max(b[2], x[2])])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.legend()
plt.show()
```
