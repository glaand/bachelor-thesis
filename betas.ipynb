```python
import numpy as np
```

```python
betas_mgpcg = np.loadtxt("experiments/2d/mgpcg/betas.dat")
betas_ml = np.loadtxt("experiments/2d/ml/betas.dat")
```

```python
# remove zeros
betas_mgpcg = betas_mgpcg[betas_mgpcg != 0]
betas_ml = betas_ml[betas_ml != 0]
```

```python
# plot the betas with scatter plot
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(np.arange(len(betas_mgpcg)), betas_mgpcg, label="MGPCG")
ax.scatter(np.arange(len(betas_ml)), betas_ml, label="ML")
ax.set_xlabel("Index")
ax.set_ylabel("Beta")
ax.set_yscale("log")
ax.legend()
plt.show()
```
