```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
```

```python
plt.style.use(['science', 'no-latex'])
```

```python
low_eigenvectors_n_cg = np.loadtxt("experiments/2d/low_eigen/n_cg_over_it.dat")
high_eigenvectors_n_cg = np.loadtxt("experiments/2d/high_eigen/n_cg_over_it.dat")
low_eigenvectors_residuals = np.loadtxt("experiments/2d/low_eigen/residuals_without_pressure_solver.dat")
high_eigenvectors_residuals = np.loadtxt("experiments/2d/high_eigen/residuals_without_pressure_solver.dat")
```

```python
# plot bar chart
mean_low_eigenvectors_n_cg = np.mean(low_eigenvectors_n_cg)
mean_high_eigenvectors_n_cg = np.mean(high_eigenvectors_n_cg)


fig, ax = plt.subplots()
bar_width = 0.35
bar1 = ax.bar(0, mean_low_eigenvectors_n_cg, bar_width, label="Low eigenvalues")
bar2 = ax.bar(1, mean_high_eigenvectors_n_cg, bar_width, label="High eigenvalues")
ax.set_ylabel("Average Number of CG iterations")
ax.set_xticks([0, 1])
ax.set_xticklabels(["Low eigenvalues", "High eigenvalues"])
ax.legend()
plt.title("Performance Analysis - Number of CG Iterations")
plt.savefig("n_cg_over_it_low_high_eigen.pdf", format="pdf")
plt.show()

```

```python
# line plot residuals
plt.figure()
plt.plot(low_eigenvectors_residuals, label="Low Eigenvectors")
plt.plot(high_eigenvectors_residuals, label="High Eigenvectors")
plt.xlabel("Time steps")
plt.ylabel("Residual")
plt.legend()
plt.title("Performance Analysis - Residual Norm")
plt.yscale("log")
plt.savefig("residuals_low_high_eigen.pdf", format="pdf")
```
