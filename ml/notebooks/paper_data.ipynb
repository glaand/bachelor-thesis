```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_file = 1003
b_data = np.loadtxt(f"eigenvectors/b_{num_file}.dat")
fig = plt.figure(figsize=(20, 20))
plt.imshow(b_data, cmap='jet', interpolation='nearest')
# remove axis
plt.axis('off')
# remove white border
plt.tight_layout()
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
# save as pdf
plt.savefig(f"paper_assets/eigenvectors/{num_file}.pdf", format='pdf')
```

```python
num_file = 10000
b_data = np.loadtxt(f"ML_data/res_{num_file}.dat")
# rotate 90 degrees to the left
b_data = np.rot90(b_data, 5)
fig = plt.figure(figsize=(20, 20))
plt.imshow(b_data, cmap='jet', interpolation='nearest')
# remove axis
plt.axis('off')
# remove white border
plt.tight_layout()
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
# save as pdf
plt.savefig(f"paper_assets/simulation_data/{num_file}.pdf", format='pdf')
```
