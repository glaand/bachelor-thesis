```python
import numpy as np
import matplotlib.pyplot as plt
import os
import scienceplots
```

```python
plt.style.use(['science', 'no-latex'])
```

```python
# plot activation function logistic sigmoid
def logistic_sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = logistic_sigmoid(x)

plt.figure(figsize=(3, 3))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-5, 5)
plt.xlim(-5, 5)
plt.grid(True)
plt.savefig("af_logistic_sigmoid.pdf", format="pdf")
plt.show()
```

```python
# plot activation function tanh
def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 100)
y = tanh(x)

plt.figure(figsize=(3, 3))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-5, 5)
plt.xlim(-5, 5)
plt.grid(True)
plt.savefig("af_tanh.pdf", format="pdf")
plt.show()
```

```python
# plot activation function relu
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 100)
y = relu(x)

plt.figure(figsize=(3, 3))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-5, 5)
plt.xlim(-5, 5)
plt.grid(True)
plt.savefig("af_relu.pdf", format="pdf")
plt.show()
```

```python
# plot activation function elu
def elu(x):
    return np.where(x > 0, x, np.exp(x) - 1)

x = np.linspace(-10, 10, 100)
y = elu(x)

plt.figure(figsize=(3, 3))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-5, 5)
plt.xlim(-5, 5)
plt.grid(True)
plt.savefig("af_elu.pdf", format="pdf")
plt.show()
```

```python
# plot activation function swish
def swish(x):
    return x / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = swish(x)

plt.figure(figsize=(3, 3))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-5, 5)
plt.xlim(-5, 5)
plt.grid(True)
plt.savefig("af_swish.pdf", format="pdf")
plt.show()
```

```python
# plot activation function mish
def mish(x):
    return x * np.tanh(np.log(1 + np.exp(x)))

x = np.linspace(-10, 10, 100)
y = mish(x)

plt.figure(figsize=(3, 3))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-5, 5)
plt.xlim(-5, 5)
plt.grid(True)
plt.savefig("af_mish.pdf", format="pdf")
plt.show()
```
